```python
# Import dependencies
import pandas as pd
# Create column names array
columns = ['target', 'cap-shape', 'cap-surface', 'cap-color', 'bruises?', 'odor', 'gill-attachment' ,'gill-spacing' , 'gill-size', 'gill-color', 'stalk-shape', 'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number', 'ring-type', 'spore-print-color', 'population', 'habitat']
# Import dataset
df = pd.read_csv('agaricus-lepiota.txt', delimiter=',', names=columns)
df.head()
```

```python
# Split data into training and testing data
from sklearn.model_selection import train_test_split
X = df.drop('target', axis=1)
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
#X_train.shape
#X_test.shape
```

```python
# Import dependencies
from pycaret.classification import *
```

```python
# Initiate the ml experiment and set up the training pipeline
experiment = setup(X_train, target=y_train)
```

```python
best_model = compare_models()
```

```python
# Select the second best model (Random Forest with AUC of 1.0) from the list
second_best_model = create_model('rf')
predictions_df = predict_model(second_best_model, X_test)
```

```python
# Extract the predicted labels and store them in a new column 'Predicted'
predictions_df['Predicted'] = predictions_df['prediction_label']

# Add the actual target values from y_test in a new column 'Actual'
predictions_df['Actual'] = y_test.values
predictions_df
```
